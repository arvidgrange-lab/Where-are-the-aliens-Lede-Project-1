```python
import pandas as pd
```

```python
with open("coords_and_counts.csv", "r", encoding="utf-8") as f_in:
    lines = f_in.readlines()

cleaned_lines = []

for line in lines:
    stripped = line.strip()

    # If the row got trapped inside outer quotes:
    if stripped.startswith('"') and stripped.endswith('"'):
        # 1. Slice off the very first and very last quote
        # 2. Turn the escaped "" back into a normal "
        stripped = stripped[1:-1].replace('""', '"')

    cleaned_lines.append(stripped + "\n")

# Write it to a fresh, healed file
with open("coords_and_counts_FIXED.csv", "w", encoding="utf-8") as f_out:
    f_out.writelines(cleaned_lines)
```

```python
df = pd.read_csv("coords_and_counts_FIXED.csv")
```

```python
#We have to get rid of the rows with null values for coordinates
df = df.drop(columns="Unnamed: 0")
```

```python
df = df.dropna()
```

```python
#We have to also get rid of these "|"
df["location"] = df["location"].str.replace("LADAKH, NEPAL | SIKKIM, INDIA | BHUTAM", "LADAKH, NEPAL")
```
