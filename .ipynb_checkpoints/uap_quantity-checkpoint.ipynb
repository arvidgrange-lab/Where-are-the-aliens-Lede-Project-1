```python
import pandas as pd
import numpy as np
import warnings, os
import re
```

```python
df = pd.read_csv("uap_coords_and_loc.csv")
```

```python
df = df.drop(columns="Unnamed: 0")
```

```python
df_clean = (
    df.groupby(df.columns.tolist(), dropna=False, sort=False)
    .size()
    .reset_index(name="duplicate_count")
)
```
