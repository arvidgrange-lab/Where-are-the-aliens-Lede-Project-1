```python
#The website has each document's sighting location on the interface, so we can scrape that.
url = "https://www.war.gov/UFO/limit/1000/?type=.pdf"
```

```python
#The site shows 15 pages of results. I tried adjusting the url page limit but no luck.
#We have to scrape every page.
```

```python
#Importing pandas and headers
import pandas as pd
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 YaBrowser/19.6.1.153 Yowser/2.5 Safari/537.36'}
```

```python
#start cooking the soup
import requests
import time
from bs4 import BeautifulSoup
```

```python
#It turns out the page uses Single Page Application.
#We'll see if we can access the hidden API.
!pip install nbconvert
!jupyter nbconvert --to script Notebook1-uap-mapping.ipynb
```

```python
from playwright.sync_api import sync_playwright

master_data = []

with sync_playwright() as p:
    # headless=False means you get to watch it happen live on your screen
    browser = p.chromium.launch(headless=False)
    page = browser.new_page()

    page.goto(url)
    
    page.wait_for_load_state("networkidle")
    time.sleep(2)

    for current_page in range(1, 16):
        print(f"Scraping Page {current_page}...")

        # 1. SCRAPE THE GLASS
        # This CSS selector grabs every table row that does NOT contain a header cell (<th>)
        data_rows = page.locator("table tr:not(:has(th))").all()

        for row in data_rows:
            # Grabs the raw text of every column cell in that row as a clean Python list
            row_data = row.locator("td").all_inner_texts()

            # Ignore empty spacer rows
            if len(row_data) > 0:
                master_data.append(row_data)

        # Exit immediately if we just finished Lap 15
        if current_page == 15:
            break

        next_page_str = str(current_page + 1)
        print(f"Searching for button: '{next_page_str}'...")

        try:
            # This looks explicitly for an HTML Link whose exact visible text is "2", "3", etc.
            next_button = page.get_by_role(
                "link", name=next_page_str, exact=True
            )
            next_button.click()

            # ASP.NET pagination does not change the URL; it does a partial page inject.
            # We force a hard 3.5-second freeze to guarantee the new data finishes loading.
            time.sleep(3.5)

        except Exception as e:
            print(f"\nCRITICAL: Failed to click page {next_page_str}. Error: {e}")
            print(
                "The scraper stopped early, but saved everything it grabbed so far."
            )
            break

    browser.close()

# Step 3: Snap into a spreadsheet
print(f"\nExtraction complete. Total individual records pulled: {len(master_data)}")

df = pd.DataFrame(master_data)
df.to_csv("PURSUE_Complete_Archive.csv", index=False)
print("Saved to disk as: PURSUE_Complete_Archive.csv")
```
