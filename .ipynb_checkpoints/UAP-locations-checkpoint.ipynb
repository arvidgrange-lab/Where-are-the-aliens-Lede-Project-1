```python
import requests
loc_data = df["location"]
master_data = []


for loc in loc_data:

    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={loc}&key=AIzaSyAk92mCwsEOncAQfElCJZ-6EvJJhMO73W4"
    response = requests.get(url)
    lats = response.json()
    master_data.append(lats)

master_data
```

```python
#Good now we have the coordinates
```

```python
df_coords = pd.DataFrame(master_data)
df_coords
```

```python
len(master_data)
```

```python
#let's figure out the structure of the nested dictionaries
len(master_data[0]["results"])
```

```python
#this is the structure
master_data[0]["results"][0]["geometry"]["location"]["lat"]
```

```python
coords = []
for index, data in enumerate(master_data):
    try:
        lat = data["results"][0]["geometry"]["location"]["lat"]
        lng = data["results"][0]["geometry"]["location"]["lng"]
        
        coords.append([lat, lng])
#Since some of the data are places like "Low Earth Orbit" and "Moon", there won't be coords for all.
    except IndexError:
        print(f"\n[!] DATA CRASH AT ROW {index}:")
        print(f"The API returned this raw response: {data}\n")
        coords.append([None, None])
```

```python
df_coords = pd.DataFrame(coords)
```

```python
df_coords["Lat"] = df_coords[0]
df_coords["Lon"] = df_coords[1]
df_coords
```

```python
df_coords = df_coords.drop(columns=[0, 1])
```

```python
df_coords
```

```python
df_coords.to_csv("uap_only_coordinates.csv")
```
